```python
def solve_jug_problem(cap_jug1, cap_jug2, goal):
    explored = set()

    def search(jug1, jug2):
        if (jug1, jug2) in explored:
            return False

        print(f"Current State: ({jug1}, {jug2})")

        if jug1 == goal:
            print("Target Achieved!")
            return True

        explored.add((jug1, jug2))

        # Rule 1: Fill Jug A
        if search(cap_jug1, jug2):
            print(f"Rule Applied: Fill A -> ({cap_jug1}, {jug2})")
            return True

        # Rule 2: Fill Jug B
        if search(jug1, cap_jug2):
            print(f"Rule Applied: Fill B -> ({jug1}, {cap_jug2})")
            return True

        # Rule 3: Empty Jug A
        if search(0, jug2):
            print(f"Rule Applied: Empty A -> (0, {jug2})")
            return True

        # Rule 4: Empty Jug B
        if search(jug1, 0):
            print(f"Rule Applied: Empty B -> ({jug1}, 0)")
            return True

        # Rule 5: Pour A to B
        pour = min(jug1, cap_jug2 - jug2)
        if search(jug1 - pour, jug2 + pour):
            print(f"Rule Applied: Pour A→B -> ({jug1 - pour}, {jug2 + pour})")
            return True

        # Rule 6: Pour B to A
        pour = min(jug2, cap_jug1 - jug1)
        if search(jug1 + pour, jug2 - pour):
            print(f"Rule Applied: Pour B→A -> ({jug1 + pour}, {jug2 - pour})")
            return True

        return False

    if not search(0, 0):
        print("No Solution Found")


cap_jug1 = int(input("Capacity A: "))
cap_jug2 = int(input("Capacity B: "))
goal     = int(input("Target in A: "))

solve_jug_problem(cap_jug1, cap_jug2, goal)
```
