```python
!pip install textblob

from textblob import TextBlob

while True:
    data = input("Write a sentence (or type exit to quit): ")
    
    if data.lower() == "exit":
        break
    
    analysis_object = TextBlob(data)
    result = analysis_object.sentiment
    
    print("Polarity Score:", result.polarity)
    print("Subjectivity Score:", result.subjectivity)
    
    if result.polarity > 0:
        print("Result: Positive\n")
    elif result.polarity < 0:
        print("Result: Negative\n")
    else:
        print("Result: Neutral\n")
```
