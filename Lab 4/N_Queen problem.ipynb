  ## ***N-Queen Problem***



```python
def n_queens(n):
    """
    N-Queen problem ko backtracking se solve karta hai.
    n: Board ka size (n x n)
    """

    board = [-1] * n       
    solution_count = [0]    

    def is_safe(row, col):
        """Check karta hai ke (row, col) position safe hai ya nahi."""
        for prev_row in range(row):
            same_col      = (board[prev_row] == col)
            same_diagonal = (abs(board[prev_row] - col) == abs(prev_row - row))
            if same_col or same_diagonal:
                return False
        return True

    def print_board():
        """Current board solution print karta hai."""
        print(f"\nSolution {solution_count[0]}:")
        for i in range(n):
            row_str = " ".join("Q" if board[i] == j else "." for j in range(n))
            print(row_str)

    def backtrack(row):
        """Recursively queens rakhta hai using backtracking."""
        if row == n:
            solution_count[0] += 1
            print_board()
            return

        for col in range(n):
            if is_safe(row, col):
                board[row] = col       
                backtrack(row + 1)    
                board[row] = -1      

   
    backtrack(0)
    print(f"\nTotal solutions for {n}-Queens: {solution_count[0]}")



n = int(input("Enter value of N: "))
n_queens(n)
```
